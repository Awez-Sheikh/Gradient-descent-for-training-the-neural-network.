```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train = x_train.reshape((-1, 28, 28, 1))
x_test = x_test.reshape((-1, 28, 28, 1))
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
def createModel(activation_function = 'relu', dropout_rate = 0.0, batch_norm = False):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation = activation_function, input_shape = (28, 28, 1)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation = activation_function))
    model.add(layers.MaxPooling2D((2, 2)))
    
    if batch_norm:
        model.add(layers.BatchNormalization())
    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation = activation_function))
    model.add(layers.Dense(10, activation = 'softmax'))
    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return model
```

```python
model_1 = createModel(activation_function = 'relu')
history_1 = model_1.fit(x_train, y_train, epochs = 10, batch_size = 64, validation_data = (x_test, y_test))
```

```python
model_2 = createModel(activation_function = 'sigmoid')
history_2 = model_2.fit(x_train, y_train, epochs = 10, batch_size = 64, validation_data = (x_test, y_test))
```

```python
model_3 = createModel(activation_function = 'tanh')
history_3 = model_3.fit(x_train, y_train, epochs = 10, batch_size = 64, validation_data = (x_test, y_test))
```

```python
def plot_history(histories, labels):
    plt.figure(figsize = (12, 6))
    for i, history in enumerate(histories):
        plt.plot(history.history['accuracy'], label = f'{labels[i]} - Training')
        plt.plot(history.history['val_accuracy'], label = f'{labels[i]} - Validation', linestyle='--')
    plt.title('Model Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
histories = [history_1, history_2, history_3]
labels = ['ReLU', 'Sigmoid', 'Tanh']
plot_history(histories, labels)
```

```python
for i, model in enumerate([model_1, model_2, model_3]):
    loss, accuracy = model.evaluate(x_test, y_test)
    print(f'Model {labels[i]} - Test accuracy: {accuracy:.4f}')
```
